```python
import pandas as pd
import numpy as np
import sqlite3
```

```python
combined_df=pd.DataFrame(columns = ["area","st","year", "occ_title","occ_group","tot_emp","h_median"])
file_list=[]
file_suffix=""
file_structure=f"./state_M{file_suffix}"
for x in range(2002, 2019):
    if x <2014 :
        file_suffix=f"{x}_dl.xls"
    else:
        file_suffix=f"{x}_dl.xlsx"        
    temp_df=pd.read_excel(file_structure+file_suffix)
    temp_df.columns = [y.lower() for y in temp_df.columns]
    temp_df["year"]=x
#   The occupational group changed names in 2012 from group to occ group
#   The old ones only had major and total defined and detailed was blank also
    if x < 2012:
        temp_df["occ_group"]=temp_df["group"]
        temp_df["occ_group"].fillna('detailed',inplace=True)
    temp_df=temp_df[["area","st","year", "occ_title","occ_group","tot_emp","h_median"]]

    print(f"year = {x}")
    print(temp_df.info())
    combined_df.append(temp_df,ignore_index=True)
    file_list.append(file_structure+file_suffix)


print(file_list)
```
